```python
data=pd.read_excel('Session 5 Samsung Market Share in China Excel.xlsx','Data')
data.head()
```

# Analyze Phone Brands Market Share in China

```python
data_new=100*data.transpose()
data_new
```

```python
data_new.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_new.plot.area(figsize=(15,8))
plt.legend(bbox_to_anchor=(1, 1))
plt.title('Mobile Phone Market in China')
plt.xlabel('Year')
plt.ylabel('Market Share in %')
```

1. The band width in the chart represent the market share in China. The band width of Samsung is shrinking from 2013 to 2015. This is a big indication of performance problem
2. The band width of the other brands are stable despite some small variation year over  year. The most striking increase in band width is Xiaomi, which increase the market share from nothing to 17% by 2015. The drop in Samsung's market share is captured by Xiaomi. This might be the main causes of Samsung's market share decline.  
3. Based on 1 and 2, the manager should allocate resources to study Xiaomi and the rest of the landscape in the Chinese market. The research should address the question about what Xiaomi is doing and why it captures the market share that Samsung seems lose. After understanding this, Samsung can take action accordingly. 

# Samsung's performace in China

```python
data_new.loc[2010:2015, ['Samsung']].plot.bar(stacked=True)
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Samsung's performance in China")
plt.xlabel('Year')
plt.ylabel('Market Share in %')
```

Samsung's market share has been decreasing significantly since year 2013. At the same time, from first graph we can see that the market share of Xiaomi has been increasing during the same period. Therefore, we will compare the market dynamics for Samsung and Xiaomi for the same period.

```python
data_new.loc[2013:2015, ['Samsung', 'Xiaomi']].plot.bar(stacked=True)
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Samsung vs Xiaomi in China")
plt.xlabel('Year')
plt.ylabel('Market Share in %')
```

There is a high chance that Samsung lost its market share to Xiaomi. This theory should be further studied and analyzed.
